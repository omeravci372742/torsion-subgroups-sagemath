```python
K = CyclotomicField(43) 
```

```python
E_189b2 = EllipticCurve([0, 0, 1, -860, 9707])
```

```python
psi_43 = E_189b2.division_polynomial(43)
```

```python
psi_43.factor()
```

```python
factors = psi_43.factor()
```

```python
len(factors)
```

```python
first = factors[0][0]
```

```python
first
```

```python
R = PolynomialRing(K, 'x')
x = R.gen()

# Coerce the polynomial to the new ring
first_K = R(first)

# Factor over the new field
factored = first_K.factor()
```

```python
factored
```

```python
root = factored[0][0].roots(multiplicities=False)[0]
```

```python
root
```

```python
E = E_189b2
```

```python
a1, a2, a3, a4, a6 = E.a1(), E.a2(), E.a3(), E.a4(), E.a6()

# Define polynomial in y
S = PolynomialRing(K, 'y')
y = S.gen()

# Plug x = root into RHS and LHS
eq = y^2 + a1*root*y + a3*y - (root^3 + a2*root^2 + a4*root + a6)

# Solve for y
y_roots = eq.roots(multiplicities=False)
```

```python
P = E(K)(root, y_roots[0])
P
```
